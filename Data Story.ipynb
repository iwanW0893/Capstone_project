# Data Story of Client database and Wrangled Expenditure

The goal of the project is to learn from the value of R&D expenditure our previous clients, and predict the value of R&D expenditure for future clients. Taking our database of client information, which includes their address, industry, the time at which we began our contact, the time at which we completed our services, we could consider if there is a pattern in these features that could give clues as to the value of the R&D expenditure. Also of note is that we complete our services for our clients on a yearly basis, therefore it is possible that we are exposing more expenditure as we gain more experience with our clients, or perhaps as we develop our methods.

With the expenditure data wrangled in the data wrangling exercise, we can see that there are:
12065 records of staff expenditure
2893 records of subcontractor expenditure
2491 records of material expenditure
509 records of software expenditure

These records evidence the sums paid by our clients for R&D projects to be completed, and it is on the basis of this expenditure that we can retrieve tax relief. These records are the most relevant to calculating future R&D expenditure, therefore we will first look at this data, and move on to the client data if we can see patterns there.

In analysing the data, I have used the median of the expenditure itself, in order to see the central tendency of the data. When looking at the apportionments, I have utilized the mean in order to consider the central tendency as well as the shape with the variance factored. 

First we will look at the distribution of the costs spent on R&D by staff members. This is a calculation of the total staff costs vs their time spent on R&D related activities. From the plot, we can see that there is a huge variance in these values, but the median is £35,585.50. This is a significant proportion of the total value of R&D costs and staff costs will be a focal point of predicting R&D expenditure.



The next graph shows how much time staff members dedicate to R&D activities. The value of claimable staff costs depends on the time dedicated to R&D related activities, and the median time spent is 30.5% of their total time at work. However this does extend down to 0.1% and upwards of 90%. There is again a huge variance in the time spent on R&D which would suggest that there will be an useful mechanism to predict what value will be derived from this staff member.


If we were to successfully predict a staff members cost to R&D, we will also need to assess how many staff members this calculation can replicate to. We will therefore need to have the data set of how many staff members have been involved in R&D projects from our records. Again we can see that there are some major projects completed with many staff members involvement, but the majority of R&D related projects are small in scale; with a median value of 5.0. The difference inferred with our first graph of R&D related costs is that we can see how many staff members the original calculation is calculated against.



Finally, we have a plot to visualise how greatly the larger projects differ to the average R&D project. A few massive projects have clearly involved individuals with very high values. They perhaps had very high total salaries as well as high time apportionments.

The subcontractor expenditure is similar to the staff expenditure. The median is £10,450.92 which is significantly smaller than the staff figure, although we can see that the subcontractor costs can be much higher, extending to £297,350.67. There is a greater variance in the subcontractor costs. 



However when it comes to the apportionment of subcontractor expenditure relevant to R&D, it is near  65%, which is the cap on subcontractor costs relevant to R&D, unless they are connected to the main company. The overall mean is 58.76%. This implies that most subcontractor costs are 65% apportioned, that they had dedicated all their time to R&D purposes, while subcontractors who do not spend all of their time on R&D are more numerous than subcontractors who are connected and are 100% apportioned. 



Similarly to the staff involved, we should look at how many subcontractors may be involved in R&D projects.
Half as many subcontractors are typically involved, at approximately 3 subcontractors per annual period in comparison to 6.7 mean staff members. We should therefore expect that each subcontractor would be more valuable per capita than the staff members involved.


With the plot of number of subcontractors and the size of their value in R&D expenditure, there are some major R&D values with very few subcontractors, and many subcontractors with very small values. This is most likely as subcontractors can be individuals or major suppliers that complete R&D activity on the company's behalf. From the data we can see that the trend is the more subcontractors the more value there is, but we can't be sure. It is likely that this would be a poor predictor.


Next we look at the material used in R&D projects. What materials involved in R&D would range massively depending on the industry. We cannot expect that many materials would be utilized for software based projects. A manufacturer or vehicle engineering company would necessarily invest massively into materials used and consumed in development. With this idea and the distribution below in mind, material means would need to be taken with more care than other types of expenditure. The median for all industries is £1287.02, whereas the largest value of subcontractors involved in a single period is £499,000.00. 



Again we can look at the distribution of the material allocation to R&D. There is far greater variance in the percentages put forward for materials than for subcontractors and staff members. 100% allocation is more likely than the other expenses, as the company would be providing expenditure specifically for R&D purposes. Note that the median allocation is far smaller at 11.5% than staff and subcontractors, as would be expected considering that apportioned costs would be derived from typically used material within the business for day to day production, and not all staff members within the business would be involved in R&D activity. The variance is more significant as evidenced by the mean skewing to 32.4%, brought up likely by 100% allocations.



The number of materials utilized is very small. The median is 1.0 and the mean is 1.7. This reflects that most frequently, materials are included due to the day to day runnings of the business; that is light, heat and power. Although this is to be expected by softwares and intangible based projects, we would expect that manufacturers and engineers to skew the mean upwards. Typically a company would not reflect their material costs individually, and would incur the costs by a supplier, and reflect the costs by the supplier rather than breaking everything down individually. Perhaps material costs could be predicted to be minimal when they are broken down by the company as they are purchasing very few materials for the purpose of the project, whereas two or three material headings listed are sourced from light,heat and power and suppliers, which would evidence the greater figures of materials. 



Next we are plotting the value of materials used in R&D projects versus the number of materials used. Indeed as the question posed by the previous plot suggested, two materials suggest greater values than other expressions of the materials used. 


The next and final costs to consider are software costs. These are the least frequent costs, however they should still be factored into our analysis and exploration of the data.

Software costs are typically very small; the median being £3544.78, and is similar to the material costs. Again there is huge variance, as the maximum value is £587,221.82.



As with the material expenditure, the software expenditure allocation to R&D is in the region of 100%, meaning that we are effectively able to isolate the cost to R&D. Software expenditure very much shares the pattern of material expenditure. Although the legislation differs in treating these heads of expenditure, they should perhaps be combined to strengthen the analysis. 


Finally the R&D costs are plotted versus the number of software. Again the same pattern emerges as for the material costs. Further analysis should be used to consider whether the software and material costs should be concatenated to create a stronger conclusion.

Although the industries typically invested in R&D dominate the others in terms of average investment, we have identified other variables which may be useful in predicting client R&D investments. These variables show variations in benefit values that would be useful to model along the industry type to deepen our prediction into R&D values.
